Project Title: Sales Data Analysis using Python

Objective:
 The objective of this project is to analyse sales data to identify trends,
patterns, and key insights that can support business decision-making.

Description:
 This project focuses on understanding sales performance across different
 regions, product categories, and time periods. By analysing the data,
 we aim to determine which areas contribute most to revenue and profitability.

Tools Used:
# Python
# Pandas (for data analysis)
# Matplotlib (for data visualisation)

 Expected Outcome:
 The analysis will provide insights into high-performing regions, top-selling
 product categories, and sales trends over time, helping businesses make
 data-driven decisions.

```python
# Import pandas library (used for working with data tables)
import pandas as pd


# Create a dictionary (this is your dataset)
# Each key is a column name and values are the data
data = {
    'Order Date': ['2023-01-10', '2023-01-15', '2023-02-05', '2023-02-20', '2023-03-12',
                   '2023-03-18', '2023-04-10', '2023-04-22', '2023-05-05', '2023-05-15'],

    # Region where the sales happened
    'Region': ['East', 'West', 'East', 'South', 'West', 'South', 'East', 'West', 'South', 'East'],

    # Category of product sold
    'Category': ['Technology', 'Furniture', 'Office Supplies', 'Technology', 'Furniture',
                 'Office Supplies', 'Technology', 'Furniture', 'Office Supplies', 'Technology'],

    # Sales amount (revenue)
    'Sales': [500, 700, 300, 450, 800, 350, 600, 750, 400, 650],

    # Profit made from each sale
    'Profit': [50, 80, 30, 40, 90, 35, 60, 85, 45, 70]
}


# Convert dictionary into a DataFrame (table format)
df = pd.DataFrame(data)


# Display the dataset
df
```

```python
# Print the shape of the dataset (rows, columns)
print("Shape of dataset:", df.shape)


# Print column names (to understand what data we have)
print("\nColumn names:")
print(df.columns)


# Print data types of each column (to check if they are correct)
print("\nData types:")
print(df.dtypes)


# Print number of missing values in each column
print("\nMissing values:")
print(df.isnull().sum())
```

```python
# Convert 'Order Date' column from text (object) to datetime format
# This allows us to perform time-based analysis (like month, year, trends)
df['Order Date'] = pd.to_datetime(df['Order Date'])

# Display first 5 rows of the dataset to check the changes
df.head()
```

```python
# Extract month name (January, February, etc.) from Order Date
df['Month'] = df['Order Date'].dt.month_name()

# Extract year (e.g., 2023) from Order Date
df['Year'] = df['Order Date'].dt.year

# Display updated dataset with new columns
df
```

```python
# Calculate total sales by summing all values in 'Sales' column
print("Total Sales:", df['Sales'].sum())

# Calculate total profit by summing all values in 'Profit' column
print("Total Profit:", df['Profit'].sum())
```

```python
# Group the dataset by 'Region'
# Then calculate total sales for each region
sales_by_region = df.groupby('Region')['Sales'].sum()

# Print the result
print(sales_by_region)
```

```python
# Create a bar chart to visualise total sales for each region
sales_by_region.plot(kind='bar', title='Sales by Region')
```

```python
# Group the dataset by 'Category'
# Then calculate total sales for each category
sales_by_category = df.groupby('Category')['Sales'].sum()

# Print the result
print(sales_by_category)
```

```python
# Create a bar chart to visualise total sales for each product category
sales_by_category.plot(kind='bar', title='Sales by Category')
```

```python
# Group the dataset by 'Month'
# Then calculate total sales for each month
monthly_sales = df.groupby('Month')['Sales'].sum()

# Print the result
print(monthly_sales)
```

```python
# Create a line chart to show sales trend over months
monthly_sales.plot(kind='line', title='Monthly Sales Trend')
```
